```python
# --- Block 4: Game Variables and Functions ---

import random

# Player stats
stats = {"energy": 100, "points": 0, "mood": "happy ğŸ˜„"}

# Motivational quotes (unique feature)
motivation_quotes = [
    "Keep grinding, bestie! Even small progress is still progress ğŸ’ª",
    "Donâ€™t forget: your vibe attracts your tribe ğŸ˜",
    "Youâ€™re doing amazing, trust the process ğŸŒˆ",
    "One more step toward your glow-up âœ¨",
    "Drink water and mind your business ğŸ’…"
]

def ShowStatus():
    print("\nğŸ“Š --- Current Stats ---")
    for key, value in stats.items():
        print(f"{key.capitalize()}: {value}")
    print("-------------------------\n")

def UpdateStats(activity):
    if activity == "study":
        stats["energy"] -= 20
        stats["points"] += 10
        stats["mood"] = "focused ğŸ“š"
    elif activity == "eat":
        stats["energy"] += 15
        stats["mood"] = "satisfied ğŸ”"
    elif activity == "play":
        stats["energy"] -= 10
        stats["points"] += 5
        stats["mood"] = "excited ğŸ®"
    elif activity == "sleep":
        stats["energy"] = 100
        stats["mood"] = "rested ğŸ˜´"
    else:
        print("âŒ Invalid activity!")

    # Keep energy within limits
    stats["energy"] = max(0, min(100, stats["energy"]))

    # Show motivational quote (unique feature)
    print("ğŸ’¬", random.choice(motivation_quotes))

def EndGame():
    print("\nğŸ Game Over! Here's your summary:")
    ShowStatus()
    print("ğŸŒŸ Thanks for playing GenZ Life Simulator, bestie! ğŸŒŸ")
```
